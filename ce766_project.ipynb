# grouping on the basis elements (corresponding gas emissions) and taking data from 2008 to 2018 and taking sum of emissions for all items(agricultural methods).

```python
plt.figure(figsize=(12,7))
plt.plot(element_df.columns,element_df.loc['CH4 emissions','Y2008':'Y2018'])
plt.ylabel('CH4 emissions in Gigagrams')
plt.title('CH4 emission in India 2008-2018')
plt.grid()
```

```python
plt.figure(figsize=(12,7))
plt.plot(element_df.columns,element_df.loc['N2O emissions','Y2008':'Y2018'])
plt.ylabel('N2O emissions in Gigagrams')
plt.title('N2O emission in India 2008-2018')
plt.grid()
```

# CO2 equivalent will be sum of above emissions in CO2 equivalent unit

```python
plt.figure(figsize=(12,7))
plt.plot(element_df.columns,element_df.loc['Emissions (CO2eq)','Y2008':'Y2018'])
plt.ylabel('CO2 eq emissions in Gigagrams')
plt.title('CO2 eq emission in India 2008-2018')
plt.grid()
```

# now grouping indian data on the basis of items(Agricultural methods) from year 2008-2018

```python
items_df=df_india.groupby(['Element','Item'])['Y2008', 'Y2009', 'Y2010', 'Y2011', 'Y2012', 'Y2013', 'Y2014', 'Y2015','Y2016', 'Y2017', 'Y2018'].sum()
```

```python
items_df
```

# Analyzing agricultural methods focused in the PPT.
(Enteric Fermentation,Manure Management,Rice Cultivation,Synthetic Fertilizers,Manure applied to Soils,Manure left on Pasture,Crop Residues)

# adding value for N2O_CO2_eq and CH4_CO2_eq for different agricultural methods for the year 2018 

```python
item=['Enteric Fermentation','Manure Management','Rice Cultivation','Synthetic Fertilizers','Manure applied to Soils','Manure left on Pasture','Crop Residues']
n2o_eq=[]
for i in item:
    if(i in items_df.loc['Emissions (CO2eq) from N2O'].index):
        n2o_eq.append(items_df.loc[('Emissions (CO2eq) from N2O',i),'Y2018'])
    else:
        n2o_eq.append(0)
ch4_eq=[]
for i in item:
    if(i in items_df.loc['Emissions (CO2eq) from CH4'].index):
        ch4_eq.append(items_df.loc[('Emissions (CO2eq) from CH4',i),'Y2018'])
    else:
        ch4_eq.append(0)
```

```python
x=2*np.arange(1,1+len(item))
```

```python
plt.figure(figsize=(18,8))
plt.bar(x-0.35,n2o_eq,width=0.7,label='N2O_eq')
plt.bar(x+0.35,ch4_eq,width=0.7,label='CH4_eq')
plt.legend()
plt.title('Comparision of CH4 and N20 emissions in India in 2018 (in CO2 equivalent)')
plt.ylabel('Emissions in Gigagrams')
plt.xticks(ticks=x,labels=item)
plt.grid()
plt.show()
```
